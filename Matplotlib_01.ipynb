### Matplotlib 기본 사용

```python
import matplotlib.pyplot as plt

plt.plot( [ 1,2,3,4] )
plt.show
```

### 스타일 지정하기

```python
plt.plot( [ 1,2,3,4], [1,4,9,16], 'ro' )
plt.axis( [ 0,6,0,20] )
plt.show()
```

### 여러 개의 그래프 그리기

```python
import numpy as np

# 200ms 간격으로 균일하게 샘플된 시간
t = np.arange(0., 5., 0.2)

# 빨간 대쉬, 파란 사각형, 녹색 삼각형
plt.plot(t, t, 'r--', t, t**2, 'bs', t,t**3, 'g^')
plt.show()
```

### Matplotlib 숫자 입력하기 

```python
### plot() 함수에 두 개의 리스트를 입력하면 순서대로 x, y값들로 인식

plt.plot( [ 1,2,3,4,], [2,3,5,10] )
plt.show
```

```python
### 레이블이 있는 데이터 사용하기

data_dict = { 'data_x' : [1,2,3,4,5], 'data_y' : [2,3,5,10,8] }

plt.plot('data_x', 'data_y', data=data_dict)
plt.show()
```

### Matplotlib 축 레이블 설정하기

```python
import matplotlib.pyplot as plt

plt.plot( [ 1,2,3,4 ], [ 2,3,5,10 ] )
plt.xlabel( 'X-Axis', labelpad=15, fontdict={'family':'serif', 'color':'b', 'weight':'bold', 'size':14} )
plt.ylabel( 'Y-Axis', labelpad=20, fontdict={'family':'fantasy', 'color':'deeppink', 'weight':'normal', 'size':'xx-large'} )
plt.show()
```

```python
### fontdict 설정

font1 = {'family' : 'serif', 'color' : 'b' , 'weight' : 'bold', 'size' : '14'}
font2 = {'family' : 'fantasy', 'color' : 'r' , 'weight' : 'normal', 'size' : 'xx-large'}

plt.plot( [1,2,3,4], [3,4,5,7] )
plt.xlabel( 'X-Axis', labelpad=15, fontdict=font1)
plt.ylabel( 'Y-Axis', labelpad=20, fontdict=font2)
```

```python
### 위치 지정하기
### xlabel() left, center, right
### ylabel() bottom, center, top
plt.plot( [1,2,3,4], [3,4,5,7] )
plt.xlabel( 'X-Axis', loc='right')
plt.ylabel( 'Y-Axis', loc='top')
plt.show()
```

### Matplotlib 범례 표시하기
##### 범례(Legend)는 그래프에 데이터의 종류를 표시하기 위한 텍스트
##### matplotlib.pyplot 모듈의 legend() 함수를 사용

```python
plt.plot([1,2,3,4], [2,3,5,11], label='Price ($)')
```
