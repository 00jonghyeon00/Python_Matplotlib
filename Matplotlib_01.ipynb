### Matplotlib 기본 사용

```python
import matplotlib.pyplot as plt

plt.plot( [ 1,2,3,4] )
plt.show
```

### 스타일 지정하기

```python
plt.plot( [ 1,2,3,4], [1,4,9,16], 'ro' )
plt.axis( [ 0,6,0,20] )
plt.show()
```

### 여러 개의 그래프 그리기

```python
import numpy as np

# 200ms 간격으로 균일하게 샘플된 시간
t = np.arange(0., 5., 0.2)

# 빨간 대쉬, 파란 사각형, 녹색 삼각형
plt.plot(t, t, 'r--', t, t**2, 'bs', t,t**3, 'g^')
plt.show()
```
